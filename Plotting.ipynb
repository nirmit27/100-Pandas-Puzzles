```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

### Creating the initial DataFrame for plotting

```python
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)

df
```

### **Puzzle 1**
Plot a **line** plot.

```python
plt.figure(figsize=(8, 4))
plt.plot(df['Name'], df['Age'], marker='o')

plt.title('Age by Name', fontdict={'fontweight': 'bold'})
plt.xlabel('Name', labelpad=5)
plt.ylabel('Age', labelpad=5)

plt.show()
```

### **Puzzle 2**
Plot a **bar** graph.

```python
plt.figure(figsize=(8, 4))
plt.bar(df['Name'], df['Age'], color='skyblue')

plt.title('Age by Name', fontdict={'fontweight': 'bold'})
plt.xlabel('Name', labelpad=5)
plt.ylabel('Age', labelpad=5)

plt.show()
```

### **Puzzle 3**
Plot a **histogram**.

```python
plt.figure(figsize=(8, 4))
plt.hist(df['Age'], bins=5, color='red')

plt.title('Age Distribution', fontdict={'fontweight': 'bold'})
plt.xlabel('Age', labelpad=5)
plt.ylabel('Frequency', labelpad=5)

plt.show()
```

### Creating a sample DataFrame

```python
data1 = {'Height': [5.1, 6.2, 5.9], 'Weight': [60, 75, 72]}
df1 = pd.DataFrame(data1)

df1
```

### **Puzzle 4**
Plot a **scatter** plot.

```python
plt.figure(figsize=(8, 4))
plt.scatter(df1['Height'], df1['Weight'], color='blue')

plt.title('Height vs. Weight', fontdict={'fontweight': 'bold'})
plt.xlabel('Height', labelpad=5)
plt.ylabel('Weight', labelpad=5)

plt.show()
```

### Creating a sample DataFrame

```python
data2 = {'Month': ['Jan', 'Feb', 'Mar'], 'Temperature_A': [
    30, 35, 40], 'Temperature_B': [25, 30, 35]}
df2 = pd.DataFrame(data2)

df2
```

### **Puzzle 5**
Plot **multiple lines** on the same graph.

```python
plt.figure(figsize=(8, 4))
plt.plot(df2['Month'], df2['Temperature_A'], marker='o', label='City A', color='red')
plt.plot(df2['Month'], df2['Temperature_B'], marker='x', label='City B', color='blue')

plt.title('Monthly Temperature', fontdict={'fontweight': 'bold'})
plt.xlabel('Month', labelpad=5)
plt.ylabel('Temperature', labelpad=5)
plt.legend()

plt.show()
```

### Creating a sample DataFrame

```python
data3 = {'Category': ['A', 'B', 'C'], 'Values': [30, 40, 30]}
df3 = pd.DataFrame(data3)

df3
```

### **Puzzle 6**
Plot a **pie** chart.

```python
plt.figure(figsize=(5, 5))
plt.pie(df3['Values'], labels=[label for label in df3['Category']], autopct='%1.0f%%', colors=['#ff9999','#66b3ff','#99ff99'])

plt.title('Category Distribution', fontdict={'fontweight': 'bold'})
plt.legend()

plt.show()
```

### Creating a sample DataFrame

```python
data4 = {'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
        'Values': [30, 35, 40, 45, 50, 55]}
df4 = pd.DataFrame(data4)

df4
```

### **Puzzle 7**
Plot a **box** plot.

```python
df4.boxplot(by='Category', figsize=(6, 4))

plt.suptitle('')
plt.title('Box Plot by Category', fontdict={'fontweight': 'bold'})
plt.xlabel('Category', labelpad=5)
plt.ylabel('Values', labelpad=5)

plt.show()
```

### **Puzzle 8**
Plot a **heatmap** using `seaborn`.

```python
plt.figure(figsize=(8, 6))
sns.heatmap(np.random.rand(10, 10), annot=True, fmt='1.1f', cmap=sns.cubehelix_palette(as_cmap=True))

plt.title("Sample Heatmap", pad=12, fontdict={'fontsize': 14})
plt.show()
```

### **Puzzle 9**
Plot a **pair plot** using `seaborn`.

```python
sns.pairplot(df4, height=4, aspect=1, x_vars=['Values'], y_vars=['Values'])

plt.title("Sample pair plot", pad=15)
plt.show()
```
